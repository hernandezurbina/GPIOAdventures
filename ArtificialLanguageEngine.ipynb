```python
import re
import random
```

```python
random.seed(42)
```

```python
CONSONANTS = list("ptkmnslr")
VOWELS = list("aeiou")
```

```python
# syllable generator (CV structure)
def generate_syllable():
    return random.choice(CONSONANTS) + random.choice(VOWELS)

# create a stable mapping so same input → same output
syllable_map = {}

def get_syllable(word):
    if word not in syllable_map:
        # assign 2–3 syllables per word
        syllable_map[word] = "".join(generate_syllable() for _ in range(random.randint(2, 3)))
    return syllable_map[word]
```

```python
def reverse_letters(word):
    return word[::-1]
```

```python
def tokenize(text):
    return re.findall(r"[a-zA-Z]+|[^a-zA-Z\s]", text)
```

```python
def to_artificial_language(text, reverse=False):
    tokens = tokenize(text)
    output = []

    for token in tokens:
        if token.isalpha():
            w = reverse_letters(token.lower()) if reverse else token.lower()
            output.append(get_syllable(w))
        else:
            # keep punctuation as-is
            output.append(token)

    return " ".join(output)
```

```python
sentence = "The lazy dog jumps over the small fox."
```

```python
print("Normal:")
print(to_artificial_language(sentence))

print("\nReversed letters version:")
print(to_artificial_language(sentence, reverse=True))
```
