```python
import requests
import json
```

```python
url = "https://v6.bvg.transport.rest/locations?poi=false&addresses=false&query=buschallee"
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    print(json.dumps(response.json(), indent=2))
else:
    print("Request failed:", response.status_code)
```
